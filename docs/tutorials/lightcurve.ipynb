# Create a Lightcurve

This notebook shows one of the simplest functionalities of the catalog, how to obtain a light curve, or a series of light curves. First, import the `get_all_lcs` function from the `slsne` package.

```python
from slsne.lcurve import get_all_lcs
```

Now we can use the built in `get_all_lcs` to get the individual light curves of SLSNe, as well as the mean evolution in a specific band. The default value of `names = None` will import all the light curves of all SLSNe and return the mean and the $\pm 1 \sigma$ range in AB magnitudes. Values of 1, 2, or 3 $\sigma$ can be specified.

```python
# Get light curves of all SLSNe in r-band
band = 'r'
(dim, mean, high), (time_samples, lightcurves) = get_all_lcs(band, return_individual=True, sigma=1)

# Get a single light curve
time_samples, r_2018lfe = get_all_lcs(band, names = '2018lfe')
```

Now we can plot the results.

```python
import matplotlib.pyplot as plt
from slsne.utils import plot_colors

# Plot individual light curves shaded.
plt.plot(time_samples, lightcurves.T, color='k', alpha=0.08, linewidth=1, zorder = 1)

# Plot the mean range in r-band
plt.fill_between(time_samples, dim, high, color = plot_colors(band), alpha = 0.4, linewidth = 0, zorder = 2)
plt.plot(time_samples, mean, color = plot_colors(band), zorder = 3)

plt.title(f'{band}-band')
plt.ylabel(f'Absolute Magnitude')
plt.xlabel('Phase [Days]')
plt.ylim(-14, -22.5)
plt.xlim(-70, 300)
plt.show();
```
